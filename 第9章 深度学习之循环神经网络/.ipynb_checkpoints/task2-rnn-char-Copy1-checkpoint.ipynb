## Char Prediction using LSTM

```python
raw_data = open('flare').read()
# 移除部分字符 '\n' 
data = raw_data.replace('\n', ' ').replace('\r', '')
print(data)
```

```python
# 删除重复字符并统计输出类别
letters = list(set(data))
num_letters = len(letters)
print(letters)
```

```python
# 建立字符到数字的索引字典
char_to_int = dict(zip(letters, [i for i in range(len(letters))]))
## 建立数字到字符的索引字典
int_to_char = dict(zip([i for i in range(len(letters))],letters))
print(int_to_char)
```

```python
import numpy as np
#滑动窗口提取数据
def extract_data(data, slide):    
    x = []
    y = []    
    ## Generating iput texts(characters)
    for i in range(len(data) - slide):
        x.append([char for char in data[i:i+slide]])
        y.append(data[i+slide:i+slide+1])        
    return x,y
#字符到数字的批量转化
def char_to_int_Data(x,y, char_to_int):    
    x_to_int = []
    y_to_int = []
    for i in range(len(x)):
        x_to_int.append([char_to_int[char] for char in x[i]])
        y_to_int.append([char_to_int[char] for char in y[i]])    
    return x_to_int, y_to_int
#数字到字符的批量转化
def int_to_char_Data(y, int_to_char):    
    y_to_char = []
    for i in range(len(y)):
        y_to_char.append([int_to_char[y[i][0]]])        
    return y_to_char
#实现输入字符文章的批量处理，输入整个字符、滑动窗口大小、转化字典
def data_preprocessing(data, slide, char_to_int):    
    char_Data = extract_data(data, slide)
    int_Data = char_to_int_Data(char_Data[0], char_Data[1], char_to_int)
    Input = int_Data[0]
    Output = list(np.array(int_Data[1]).flatten())
    Input_RESHAPED = np.array(Input).reshape(len(Input), slide, 1)
    return Input_RESHAPED, Output
```

```python
#将输入文本转化为可用于模型训练的数字序列（包含输入序列与输出结果）
DATA_SET = data_preprocessing(data, 20, char_to_int)

X = DATA_SET[0]
y = DATA_SET[1]
```

```python
#数据分离
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=10)
```

```python
# 把输出结果转化为多分类的类别索引数组[0,0,0,..1..0]
from keras.utils import to_categorical
y_train_category = to_categorical(y_train, num_classes=num_letters)
y_test_category = to_categorical(y_test, num_classes=num_letters)
print(y_test_category[1])
```

```python
print(X_train.shape,y_train_category.shape)
```

```python
print(X_train[1])
```

```python
# 建立一个两个隐藏层的lstm
from keras.models import Sequential
from keras.layers import LSTM,Dense

## Sequential one by one
model = Sequential()

## LSTM Layer 1: Consists of 256 Neurons. To connect with the second layer of LSTM, return_sequences = True
model.add(LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
## LSTM Layer 2: Consists of 256 Neurons in one RNN Layer 
model.add(LSTM(128))
model.add(Dense(num_letters, activation="sigmoid"))

## Compiling Model
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# 建立一个两个隐藏层的lstm
from keras.models import Sequential
from keras.layers import LSTM,Dense

## Sequential one by one
model = Sequential()
model.add(LSTM(128, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(num_letters, activation='softmax')) #输出用一个softmax，来分类，维度就是26，预测是哪一个字母
 
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

```

```python
## Fitting Model 
model.fit(X_train, y_train_category, batch_size=1000, epochs=3)
```

```python
y_test_predict = model.predict_classes(X_test)
print(y_test_predict,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_test = accuracy_score(y_test,y_test_predict)
print(accuracy_test)
```

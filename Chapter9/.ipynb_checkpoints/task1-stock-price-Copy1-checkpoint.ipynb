```python
import pandas as pd
import numpy as np

data = pd.read_csv('zgpa_train.csv')
data.head()
```

```python
price = data.loc[:,'close']
num_data = price.shape[0]
print(num_data)
```

```python
X = price[0:num_data-1]
y = price[1:num_data]
print(X.shape,y.shape)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
fig1 = plt.figure(figsize=(7,5))
plt.plot(y)
plt.show()
```

```python
y.shape
```

```python
#normalize the data
X_norm = X/max(X)
y_norm = y/max(y)
X_train = X_norm
y_train = y_norm
plt.plot(y_train)
```

```python
X_train = np.array(X_train).reshape(num_data-1,1,1)
print(X_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,SimpleRNN
model = Sequential()
model.add(SimpleRNN(units=5,input_shape = (730,1),activation='relu'))
model.add(Dense(units=1,activation='linear'))
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.summary()
```

```python
X_train.shape
```

```python
model.fit(X_train.reshape(1,730,1),y_train,batch_size=30,epochs=200)
```

```python
#make prediction
y_train_predict = model.predict(X_train)
y_train_predict = y_train_predict*max(y)
```

```python
fig3 = plt.figure(figsize=(7,5))
real_price = plt.plot(y,label='real_price')
predict_price = plt.plot(y_train_predict,label='predict_price')
plt.legend()
plt.show()
```

```python
data_test = pd.read_csv('zgpa_test.csv')
data_test.head()
price_test = data_test.loc[:,'close']
num_test = price_test.shape[0]
X_test = price_test[0:num_test-1]
y_test = price_test[1:num_test]
X_test = X_test/max(X)
X_test = np.array(X_test).reshape(num_test-1,1,1)
y_predict_test = model.predict(X_test)
y_predict_test = y_predict_test*max(y)
```

```python
fig4 = plt.figure(figsize=[7,5])
plt.plot(y_test,label='real price(test)')
plt.plot(y_predict_test,label='predict price(test)')
plt.legend()
plt.show()
```

```python
y_predict_test.shape
```

```python
result = np.concatenate((np.array(y_test).reshape(-1,1),y_predict_test),axis = 1)
result = pd.DataFrame(result,columns=['real_price','predicted_price'])
result.head()
result.to_csv('zgpa_predict2.csv')
```
